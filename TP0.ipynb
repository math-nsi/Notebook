```python
import matplotlib.image as mpimg # On importe le module pour les images
import numpy as np
import matplotlib.pyplot as plt #on importe le module pour l'affichage
img = mpimg.imread("images/cascade.png") # On charge l'image en mémoire
if img.dtype == np.float32: # Si le résultat n'est pas un tableau d'entiers
    img = (img * 255).astype(np.uint8)
hauteur = img.shape[0]
largeur = img.shape[1]
print("hauteur en pixels : "+str(hauteur))
print("largeur en pixels : "+str(largeur))
print("pixel en haut à gauche : "+str(img[0][0]))
print("pixel en haut à droite : "+str(img[0][largeur-1]))
plt.imshow(img)
plt.show()
```
